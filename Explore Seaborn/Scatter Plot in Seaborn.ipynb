```python
# ==========================================
#           Seaborn Scatter Plot 
# ==========================================
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
sns.scatterplot(data = df, 
    x = "bill_length_mm", 
    y = "bill_depth_mm",
    hue="sex",
    )
plt.show()
```

```python
df = sns.load_dataset('penguins').head(30)
sns.scatterplot(data = df, 
    x = "bill_length_mm", 
    y = "bill_depth_mm",
    hue="sex",
    )
plt.show()
```

```python
df = sns.load_dataset('penguins')
sns.scatterplot(data = df, 
    x = "bill_length_mm", 
    y = "bill_depth_mm",
    hue="sex",
    style="sex",
    sizes=(4, 100)
    )
plt.show()
```

```python
df = sns.load_dataset('penguins')
sns.scatterplot(data = df, 
    x = "bill_length_mm", 
    y = "bill_depth_mm",
    hue="sex",
    style="sex",
    sizes=(4, 100),
    palette="icefire"
    )
plt.show()
```

```python
m = {"Male":"*", "Female":"o"}
df = sns.load_dataset('penguins')
sns.scatterplot(data = df, 
    x = "bill_length_mm", 
    y = "bill_depth_mm",
    hue="sex",
    style="sex",
    sizes=(4, 100),
    palette="icefire",
    markers=m
    )
plt.show()
```
