```python
# ================================================
#                Bar Plot in Seaborn
# ================================================
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("penguins")
df
```

```python
sns.barplot(x = df['island'], y = df['bill_length_mm'])
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex")
plt.show()
```

```python
order_1 = ["Dream", "Biscoe", "Torgersen"]
sns.barplot(x = "island", y = "bill_length_mm", 
    data = df, hue = "sex", order=order_1)
plt.show()
```

```python
order_1 = ["Dream", "Biscoe", "Torgersen"]
sns.barplot(x = "island", y = "bill_length_mm", 
    data = df, hue = "sex", order=order_1,
    hue_order=["Female", "Male"]
    )
plt.show()
```

```python
order_1 = ["Dream", "Biscoe", "Torgersen"]
sns.barplot(x = "island", y = "bill_length_mm", 
    data = df, hue = "sex", order=order_1, hue_order=["Female", "Male"],
    ci = 20
    )
plt.show()
```

```python
order_1 = ["Dream", "Biscoe", "Torgersen"]
sns.barplot(x = "island", y = "bill_length_mm", 
    data = df, hue = "sex", order=order_1, hue_order=["Female", "Male"],
    ci = 20,
    n_boot=50
    )
plt.show()
```

```python
order_1 = ["Dream", "Biscoe", "Torgersen"]
sns.barplot(x = "island", y = "bill_length_mm", 
    data = df, hue = "sex", order=order_1, hue_order=["Female", "Male"],
    ci = 20,
    n_boot=50,
    palette='icefire'
    )
plt.show()
```

```python
order_1 = ["Dream", "Biscoe", "Torgersen"]
sns.barplot(x = "island", y = "bill_length_mm", 
    data = df, hue = "sex", order=order_1, hue_order=["Female", "Male"],
    ci = 20,
    n_boot=50,
    palette='icefire',
    saturation=0.5,
    errcolor='r',
    errwidth=15,
    capsize=0.5
    )
plt.show()
```
