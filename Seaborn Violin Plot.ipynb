```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.tail()
```

```python
sns.violinplot(x='day', y='total_bill', data=df)
```

```python
sns.violinplot(x='day', y='total_bill',hue='time', data=df)
```

```python
sns.violinplot(x='day', y='total_bill', data=df, linewidth=5)
```

```python
sns.violinplot(x='day', y='total_bill', data=df, linewidth=3, palette='icefire')
```

```python
sns.violinplot(x='day', y='total_bill', data=df, saturation=0.5)
```

```python
sns.violinplot(x='day',
               y='total_bill',
               data=df, 
               saturation=0.5,
               hue='sex',
               split=True
               )
```

```python
sns.violinplot(x='time',
               y='total_bill',
               data=df, 
               saturation=0.5,
               hue='sex',
               split=True,
               order=['Dinner', 'Lunch']
               )
```

```python
sns.violinplot(x='time',
               y='total_bill',
               data=df, 
               saturation=0.5,
               hue='sex',
               split=True,
               order=['Dinner', 'Lunch'],
               inner='quarts'
               )
```

```python
sns.violinplot(x='time',
               y='total_bill',
               data=df, 
               saturation=0.5,
               hue='sex',
               split=True,
               order=['Dinner', 'Lunch'],
               inner='stick'
               )
```

```python
sns.violinplot(x='time',
               y='total_bill',
               data=df, 
               saturation=0.5,
               hue='sex',
               split=True,
               order=['Dinner', 'Lunch'],
               inner='point'
               )
```
